롤러코스터 유형 속성에는 몇 가지 유형이 있을까? 어떤 유형이 가장 많을까?

```python
x1 = df['rollercoaster_type'].unique()
x1
```

image.png

```python
x2 = df['rollercoaster_type'].nunique()
x2
```

image.png

```python
type_count = df['rollercoaster_type'].value_counts()
type_count
```

image.png

```python
type_count.plot.bar()
plt.show()
```

그래프의 전체 크기와 막대 너비 조절

image.png

```python
plt.figure(figsize=(10,6))
type_count.plot.bar(width = 0.9)
plt.xlabel('rollercoaster_type')
plt.ylabel('count')
plt.show()
```

* 그래프 위에 숫자 표기

image.png

```python
plt.figure(figsize=(10,6))
type_count.plot.bar(width = 0.9)
plt.xlabel('rollercoaster_type')
plt.ylabel('count')
plt.text(0,22.5,22)
plt.text(5,7.5,7)
plt.text(13,4.5,4)
plt.show()
```

규칙을 찾아 반복문 적용

```python
plt.figure(figsize=(10,6))
type_count.plot.bar(width = 0.9)
plt.xlabel('rollercoaster_type')
plt.ylabel('count')
for i in range(len(x1)):
    plt.text(i,
             type_count[i]+0.5,
             type_count[i],
             va='center',
             ha='center')
plt.show()
```

Q. 최고 속도, 탑승 시간, 롤러코스터 길이, 최고 낙하 높이 속성 중 흥미도 속성과 가장 관련이 깊은 속성은 무엇일까?

=> 상관관계 분석

image.png

```python
cr = df.corr()
cr
```

* 상관관계표

image.png

* 흥미도와 관련된 결과

```python
cr2 = cr['excitement']
cr2
```

```python
cr3 = cr2.sort_values(ascending=False)
cr3
```

* 막대 그래프로 시각화

```python
cr3.plot.bar()
for i in range(len(cr2)):
    plt.text(i, cr3[i]+0.05, round(cr3[i], 2),
             va='center',
             ha='center')
```
